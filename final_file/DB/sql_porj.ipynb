```python
ps = """
create table power_sales2(송변전손실_배전량B FLOAT,
                          배전손실_판매량C FLOAT  PRIMARY KEY,
                          손실량_B_C FLOAT,
                          손실률 FLOAT
)
"""
conn.execute(ps)
```

```python
sql = """
ALTER TABLE `powers1` ADD CONSTRAINT `FK_powers1_power_A` FOREIGN KEY (`송전단_전력량A`) REFERENCES `power_distribution2` (`송전단_전력량A`);
"""
conn.execute(sql)
```

```python
sql = """
ALTER TABLE `powers1` ADD CONSTRAINT `FK_powers1_ps_C` FOREIGN KEY (`배전손실_판매량C`) REFERENCES `power_sales2` (`배전손실_판매량C`);
"""
conn.execute(sql)
```
