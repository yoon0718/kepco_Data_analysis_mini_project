```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings 
import datetime as dt
warnings.filterwarnings(action='ignore')
plt.rcParams['font.family'] = 'Gulim'
pd.options.display.float_format = '{:.5f}'.format
```

```python
지역전력판매_2018 = pd.read_csv('2018 지역별 전력판매량.csv',encoding='cp949')
지역전력판매_2019 = pd.read_csv('2019 지역별 전력판매량.csv',encoding='cp949')
지역전력판매_2020 = pd.read_csv('2020 지역별 전력판매량.csv',encoding='cp949')
지역전력판매_2021 = pd.read_csv('2021 지역별 전력판매량.csv',encoding='cp949')
지역전력판매_2018
```

```python
지역전력발전 = pd.read_excel('2018~2021 지역별 발전량.xlsx')
```

```python
지역전력발전['합계'] = 지역전력발전.sum(axis = 1)
지역전력발전.index = 지역전력발전.연도
지역전력발전 = 지역전력발전.iloc[:,1:].transpose()
지역전력발전 = 지역전력발전.reset_index()
지역전력발전 = 지역전력발전[['index',2018,2019,2020,2021]]
지역전력발전.columns = ['지역',2018,2019,2020,2021]
지역전력발전
```

```python
# 지역 동일하게 맞추기
지역전력판매_2018 = 지역전력판매_2018.loc[지역전력판매_2018['구분'].isin(지역전력발전['지역']),:]
지역전력판매_2019 = 지역전력판매_2019.loc[지역전력판매_2019['구분'].isin(지역전력발전['지역']),:]
지역전력판매_2020 = 지역전력판매_2020.loc[지역전력판매_2020['구분'].isin(지역전력발전['지역']),:]
지역전력판매_2021 = 지역전력판매_2021.loc[지역전력판매_2021['구분'].isin(지역전력발전['지역']),:]
```

```python
지역전력발전 = 지역전력발전.sort_values(by = ['지역']).reset_index().iloc[:,1:]
지역전력발전_지역 = 지역전력발전.iloc[:-1,:]
```

```python
#2018~2021 지역별 발전량 
지역전력발전_지역
```

```python
합계_2018_2021 = pd.concat([지역전력판매_2018.합계,지역전력판매_2019.합계,지역전력판매_2020.합계,지역전력판매_2021.합계],axis = 1)
합계_2018_2021.columns = [2018,2019,2020,2021]
합계_2018_2021 = pd.concat([합계_2018_2021,지역전력판매_2018.구분],axis = 1)
합계_2018_2021= 합계_2018_2021[['구분',2018,2019,2020,2021]]
합계_2018_2021 = 합계_2018_2021.reset_index().iloc[:,1:]
합계_2018_2021_지역 = 합계_2018_2021.iloc[:-1,:]
```

```python
#2018~2021 지역별 판매량
합계_2018_2021_지역
```

```python
#2021년 지역별 발전량, 지역별 전기 판매량 비교(2018~2021 비슷한 양상)
fig, ax = plt.subplots(figsize=(15,5))
bar_width = 0.13
index = np.arange(17)
plt.bar(index , 지역전력발전_지역.iloc[:,4] , label="2021 지역별 발전량", width=0.3)
plt.bar(index+ 0.3 , 합계_2018_2021_지역.iloc[:,4] , label = "2021 지역별 전기 판매량", width=0.3)
plt.xticks(np.arange(bar_width, 17 + bar_width, 1), 지역전력발전_지역.지역)
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.legend()
plt.show()
```

```python
#연도별 발전량 , 판매량 차이 (총 손실율) 그래프
fig, ax = plt.subplots(figsize=(15,5))
plt.plot(지역전력발전.columns[1:],  지역전력발전.iloc[17,1:],label = '전국 발전량')
plt.plot(합계_2018_2021.columns[1:], 합계_2018_2021.iloc[17,1:],label = '전력 판매량')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.legend()
plt.show()
```

```python
한전부채 = pd.read_excel('한전부채.xlsx')
```

```python
한전부채
```

```python
한전부채.columns = 한전부채.iloc[0,:]
한전부채 = 한전부채.iloc[2:,:]
한전부채 = 한전부채.reset_index()
한전부채 = 한전부채.iloc[:,3:]
한전부채.columns = 한전부채.columns.astype(int)
```

```python
한전부채
```

```python
ratio = pd.DataFrame()
ratio['부채비율'] = 한전부채.iloc[5,:]/(한전부채.iloc[5,:]+한전부채.iloc[0,:])*100
ratio['자산비율'] = 한전부채.iloc[0,:]/(한전부채.iloc[5,:]+한전부채.iloc[0,:])*100
ratio = ratio.transpose()
```

```python
ratio
```

```python
#한국전력공사 연도별 총 부채, 자산 그래프
fig, ax = plt.subplots(figsize=(20,5))
plt.bar(한전부채.columns ,  한전부채.iloc[5,:] ,label = '연도별 한국전력공사 부채')
plt.bar(한전부채.columns , 한전부채.iloc[0,:],label = '연도별 한국전력공사 자산' , bottom = 한전부채.iloc[5,:] )
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.xticks(한전부채.columns)
plt.ylabel('(단위 : 억원)',rotation = 'horizontal', loc='top')
for i in range(len(한전부채.columns)):
    plt.text(한전부채.columns[i] ,한전부채.iloc[5,i] ,str(round(ratio.iloc[0,i],1))+'%' ,fontweight = 'heavy', fontsize = 12, va = 'top', ha='center',color = 'w')
    plt.text(한전부채.columns[i] ,한전부채.iloc[0,i] ,str(round(ratio.iloc[1,i],1))+'%',fontweight = 'heavy',fontsize = 12,va = 'top' ,ha='center')
plt.legend()
plt.show()
```

```python
지역전력판매_2021 = 지역전력판매_2021.loc[지역전력판매_2021['구분'].isin(지역전력발전['지역']),:]
지역전력판매_2021
```

```python
지역전력판매_2021_산업용 = 지역전력판매_2021.iloc[:,[0,-2]]
지역전력판매_2021_산업용.columns = ['구분','산업용판매량']
지역전력판매_2021_산업용 = 지역전력판매_2021_산업용.reset_index().iloc[:,1:]
지역전력판매_2021_산업용 = 지역전력판매_2021_산업용.iloc[:-1,:]
```

```python
지역전력판매_2021_산업용
```

```python
#2021 산업용 판매단가 = 105.48원
#인상금액 = 24원(부채/30년기준,손해금액포함)
#인상하지않는 지역의 인상금액 = 24*7/10 = 16.8원
#인상 후 판매단가 = 146.28
지역전력판매_2021_산업용.산업용판매량 = 지역전력판매_2021_산업용.산업용판매량.astype('float')
```

```python
#MKW->KW변환
지역전력판매_2021_산업용['인상후판매금액'] =  지역전력판매_2021_산업용.산업용판매량*1000
지역전력판매_2021_산업용.산업용판매량 = 지역전력판매_2021_산업용.산업용판매량 * 1000
```

```python
지역전력판매_2021_산업용
```

```python
지역전력판매_2021_산업용.iloc[[1,4,5,6,8,9,10,13,14,16],2] = 지역전력판매_2021_산업용.iloc[[1,4,5,6,8,9,10,13,14,16],2]*146.28
지역전력판매_2021_산업용.iloc[[0,2,3,7,11,12,15],2] = 지역전력판매_2021_산업용.iloc[[0,2,3,7,11,12,15],2] * 105.48
지역전력판매_2021_산업용['인상전판매금액'] = 지역전력판매_2021_산업용.산업용판매량 * 105.48
지역전력판매_2021_산업용.loc['합계']=지역전력판매_2021_산업용.sum(axis=0) 
지역전력판매_2021_산업용.loc['합계','구분'] = '합계'
지역전력판매_2021_산업용 = 지역전력판매_2021_산업용.reset_index().iloc[:,1:]
지역전력판매_2021_산업용
```

```python
fig, ax = plt.subplots(figsize=(20,10))
bar_width = 0.13
index = np.arange(18)
plt.bar(index , 지역전력판매_2021_산업용.iloc[:,3] , label="인상전판매금액", width=0.3)
plt.bar(index+0.3 , 지역전력판매_2021_산업용.iloc[:,2] , label = "인상후판매금액", width=0.3)
plt.xticks(np.arange(bar_width, 18 + bar_width, 1), 지역전력판매_2021_산업용.구분)
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.legend()
plt.show()
```
